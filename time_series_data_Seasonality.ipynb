```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data={
    "date":pd.to_datetime(["2022-01-01","2022-02-01","2022-03-01","2022-04-01","2022-05-01","2022-06-01","2022-07-01","2022-08-01","2022-09-01"]),
    "value":[15,10,25,12,20,9,16,2,18]
}
df=pd.DataFrame(data)
df.set_index("date",inplace=True)
plt.figure(figsize=(10,6))
plt.plot(df.index,df["value"],marker=' ')
plt.xlabel('Date')
plt.ylabel("Value")
plt.title("Time series_seasonality")
plt.grid(True)
plt.show()
```
