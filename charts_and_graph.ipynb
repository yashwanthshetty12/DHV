```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[2,3,4,5]
plt.plot(x,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Line plot")
plt.show()

```

```python
x=[1,2,3,4,5]
y=[2,4,5,6,7]
colors=["r","y","b","g","b"]
sizes=[100,200,300,400,500]
plt.scatter(x,y,c=colors,s=sizes,alpha=0.9,marker="*",edgecolor="g")
plt.title("scatter",fontsize=20)
plt.xlabel("X axis",fontsize=20)
plt.ylabel("Y axis",fontsize=20)
plt.show()
```

```python
import numpy as np
x=np.array([15,15,15,15])
mylabels=["apple","bananas","cherries","dates"]
myexplode=explode=[0.15,0,0,0]
colors=["b","w","b","w"]
plt.pie(x,labels=mylabels,colors=colors,explode=myexplode)
plt.show()
```

```python
X1=[3,3,4,5,6]
plt.hist(X1,color="r",edgecolor="g",bottom=1,histtype="bar",orientation="vertical",label="python")
plt.axvline(5,color="g")
plt.legend()
plt.show()


```
