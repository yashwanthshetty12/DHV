```python
import numpy as np
import matplotlib.pyplot as plt
x_pos=0
y_pos=0
a_pos=0
b_pos=0
x_direct=-1
x_direct=-1
a_direct=1
b_direct=1
fig,ax=plt.subplots(figsize=(13,9))
ax.quiver(x_pos,y_pos,x_direct,y_direct,color='r',scale=5)
ax.quiver(a_pos,b_pos,a_direct,b_direct,color='b',scale=5)
ax.set_title('Quiver plot with one arrow')
plt.show()
```

```python

x = np.arange(0, 2.2, 0.2)
y = np.arange(0, 2.2, 0.2)

X, Y = np.meshgrid(x, y)
u = np.cos(X)*Y
v = np.sin(Y)*Y

fig, ax = plt.subplots(figsize =(14, 8))
ax.quiver(X, Y, u, v)

ax.xaxis.set_ticks([])
ax.yaxis.set_ticks([])
ax.axis([-0.3, 2.3, -0.3, 2.3])
ax.set_aspect('equal')

plt.show()
```
