```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Age_group': ['<21', '21-24','24-27','27-30','30-33','33-36','36-39','57-60','>60'],
    'Totle_no_of_loans': [9,310,511,4000,4568,5698,8209,788,6],
    'No_of_badloans': [2,14,20,172,169,188,197,9,0],
    'NO_of_goodloans': [7,296,491,3828,4399,5510,8012,779,6],
    'Percentage_badloans': ['22%','4.5%','3.9%','4.3%','3.7%','3.3%','2.4%','1.1%','0.0%'],
    'Percentage_goodloans': ['77.8%','95.5%','96.1%','95.7%','96.3%','96.7%','97.6%','98.9%','100%']
}
```

```python
df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Age_group'], df['Totle_no_of_loans'], color='blue')
plt.xlabel('Age Group')
plt.ylabel('Total Number of Loans')
plt.title('Total Number of Loans by Age Group')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
percent_bad_loans = [float(percent.strip('%')) for percent in df['Percentage_badloans']]
plt.bar(df['Age_group'], percent_bad_loans, color='red')
plt.xlabel('Age Group')
plt.ylabel('Percentage of Bad Loans')
plt.title('Percentage of Bad Loans by Age Group')
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Sample data
loan_categories = ['Good Loans', 'Bad Loans']
loan_types = ['Type 1', 'Type 2']
loan_amounts_good = [[50000, 30000],
                     [20000, 40000]]  # Amount of good loans for each type
loan_amounts_bad = [[25000, 15000],
                    [10000, 20000]]   # Amount of bad loans for each type

# Create figure and 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Positions for the bars
xpos = range(len(loan_categories))
ypos = range(len(loan_types))

# Create 3D bars for good loans
for i in range(len(loan_categories)):
    for j in range(len(loan_types)):
        ax.bar3d(xpos[i], ypos[j], 0, 1, 1, loan_amounts_good[i][j], color='green', alpha=0.7)

# Create 3D bars for bad loans
for i in range(len(loan_categories)):
    for j in range(len(loan_types)):
        ax.bar3d(xpos[i], ypos[j], 0, 1, 1, loan_amounts_bad[i][j], color='red', alpha=0.7)

# Set labels and title
ax.set_xlabel('Loan Category')
ax.set_ylabel('Loan Type')
ax.set_zlabel('Loan Amount')
ax.set_xticks(range(len(loan_categories)))
ax.set_yticks(range(len(loan_types)))
ax.set_xticklabels(loan_categories)
ax.set_yticklabels(loan_types)
ax.set_title('3d Comparison of Good and Bad Loans')

# Show plot
plt.show()
```
